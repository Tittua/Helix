```python
# HYPERPARAMETERS
ARCHITECTURE = "mask_rcnn_R_101_FPN_3x"
CONFIG_FILE_PATH = f"COCO-InstanceSegmentation/{ARCHITECTURE}.yaml"
MAX_ITER = 35000
EVAL_PERIOD = 200
BASE_LR = 0.001
NUM_CLASSES = 498

# OUTPUT DIR
OUTPUT_DIR_PATH = os.path.join(
    DATA_SET_NAME, 
    ARCHITECTURE, 
    datetime.now().strftime('%Y-%m-%d-%H-%M-%S')
)

os.makedirs(OUTPUT_DIR_PATH, exist_ok=True)
```

```python
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file(CONFIG_FILE_PATH))
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(CONFIG_FILE_PATH)
cfg.DATASETS.TRAIN = (TRAIN_DATA_SET_NAME,)
cfg.DATASETS.TEST = (TEST_DATA_SET_NAME,)
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64
cfg.TEST.EVAL_PERIOD = EVAL_PERIOD
cfg.DATALOADER.NUM_WORKERS = 2
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.INPUT.MASK_FORMAT='bitmask'
cfg.SOLVER.BASE_LR = BASE_LR
cfg.SOLVER.MAX_ITER = MAX_ITER
cfg.MODEL.ROI_HEADS.NUM_CLASSES = NUM_CLASSES
cfg.OUTPUT_DIR = OUTPUT_DIR_PATH
```

```python
def image_segmentation(x):
    img = cv2.imread(x)
    outputs = predictor(img)
    
    visualizer = Visualizer(
        img[:, :, ::-1],
        metadata=metadata, 
        scale=0.8, 
        instance_mode=ColorMode.IMAGE_BW
    )
    out = visualizer.draw_instance_predictions(outputs["instances"].to("cpu"))
    cv2_imshow(out.get_image()[:, :, ::-1]) 
    return outputs
```

```python
#This function takes the image and display the detection and also their nutritious values
def combine_model(x):
  output_=[]
  output_.append(image_classification(x))
  print('\n')
  img_seg=image_segmentation(x)
  print('\n')
  output_.append(object_detection_model(x))
  print_detection(output_,img_seg)
  
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
def image_classification(x):
  img = image.load_img(x, target_size=(224, 224))
  x=image.img_to_array(img)    #Preprocessing
  x=x/255   #Scaling
  x=np.expand_dims(x, axis=0)
  preds = classification_model.predict(x)
  preds=np.argmax(preds, axis=1)
  preds
  results2=log[log['index']==(preds[0])]['0']
  print('Classification=',results2)
  return results2
```

# Output

```python
food_value=pd.read_csv('/content/drive/MyDrive/Food_Recognition/csv/food_value.csv')
segment_data=pd.read_csv('/content/drive/MyDrive/Food_Recognition/csv/segmentation_data.csv')
classification_data=pd.read_csv('/content/drive/MyDrive/Food_Recognition/csv/food_value_classification.csv')
segment_data.drop(['Unnamed: 0.1','Unnamed: 0','name_rechabel'],axis=1,inplace=True)
```

```python
def classification_results(result):
#Classification
  temp=str(result[0].values[0]).lower()
  output=classification_data[classification_data['food']==temp]
  return output
```

```python
def segmentation_results(seg):
  try:
    for i in range(len(seg['instances'].pred_classes)):
      id=str(str(seg['instances'].pred_classes[i]).split('(')[1]).split(',')[0]
      print(segment_data[segment_data['id']==int(id)],'\n')
    
  except:
    print('No detection')
  return 
```

```python
#This function helps to filter the the detected items (When detecting multiple itmes a addition 's' was 
#added at the end of the food name this problem is taken care using this function')
def yolo_results(result):
  try:
    first=str(str(str(result[1]).split(':')[1]).split('\n')[0]).split(',')[0]
    if int(first.split()[1:][0])>1:
      out=first.split()[1:][1][:-1]
      print(food_value[food_value['food']==out])
    else:
      out=first.split()[1:][1]
      print(food_value[food_value['food']==out])
  except:
    print('No detection')

  if len(str(str(str(result[1]).split(':')[1]).split('\n')[0]).split(','))>1:
    names=str(str(str(result[1]).split(':')[1]).split('\n')[0]).split(',')[1:]
    for x in names:
      if int(x.split()[0])>1:
        output=x.split()[1][:-1]
        print(food_value[food_value['food']==output])
      else:
        output=x.split()[1]
        print(food_value[food_value['food']==output])
```

```python
def print_detection(result,seg):
  print('Segmentation \n')
  segmentation_results(seg)
  print('Classification Results \n')
  print(classification_results(result),'\n')
  print('object detection Results \n')
  yolo_results(result)
```
